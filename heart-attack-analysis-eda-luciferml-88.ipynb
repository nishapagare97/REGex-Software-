<h1 style="text-align:center;font-weight: bold;">Symptoms</h1>
    <h3 style="text-align:center;font-weight: bold">Symptoms of a heart attack include:

<ul>
<li>Discomfort, pressure, tightness, squeezing, or pain in your chest or arm or below your breastbone</li>
<li>Discomfort that goes into your back, jaw, throat, or arm</li>
<li>Fullness, indigestion, or a choking feeling (it may feel like heartburn)</li>
<li>Sweating, upset stomach, vomiting, or dizziness</li>
</ul>

```python
params = {'C': 0.23357214690901212, 'penalty': 'l1', 'solver': 'saga'}
classifier =  Classification(predictor = 'lr', 
                                               params=params, lda = 'y',smote = 'y')
classifier.fit(features, labels)
result = classifier.result()
accuracy_scores[result['Classifier']] = result['Accuracy']
```

```python
params = {'C': 0.004832930238571752, 'gamma': 0.7, 'kernel': 'sigmoid'}
classifier =  Classification(predictor = 'svm', 
                                               params=params, lda = 'y',smote = 'y')
classifier.fit(features, labels)
result = classifier.result()
accuracy_scores[result['Classifier']] = result['Accuracy']
```

```python
params = {'algorithm': 'auto', 'n_jobs': 1, 'n_neighbors': 20, 'weights': 'uniform'}
classifier =  Classification(predictor = 'knn', 
                                               params=params, lda = 'y',smote = 'y')
classifier.fit(features, labels)
result = classifier.result()
accuracy_scores[result['Classifier']] = result['Accuracy']
```
